#--> **Randomly** generated **NumPy Arrays**

```python
#Create an array of the given shape and populate it with random samples from a "UNIFORM DISTRIBUTION" over [0,1)
# UNIFORM DISTRIBUTION: meaning all those numbers between [0,1) has the same likelihood of being picked

# np.random.rand(shape)

np.random.rand(4)
```

```python
np.random.rand(4,5)
```

```python
# returns samples from the "STANDARD NORMAL DISTRIBUTION"
# np.random.randn(shape)

np.random.randn(5)
```

```python
np.random.randn(4,5)
```

```python
# np.random.randint(start, stop, shape)

np.random.randint(0,101,5)
```

```python
np.random.randint(0,101,(4,5))
```

```python
# Seed --> to get a particular set of random numbers

np.random.seed(42)
np.random.randint(0,101,5)
```

```python
np.random.seed(42)
np.random.rand(4)
```
