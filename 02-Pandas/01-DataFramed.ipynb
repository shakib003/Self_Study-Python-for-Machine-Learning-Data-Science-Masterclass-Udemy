# DataFrames

**DataFrame:** A Pandas DataFrame consists of multiple Pandas Series that share index values.



```python
import numpy as np
import pandas as pd
```

----
### Creating a DataFrame from Python Object
----

```python
np.random.seed(101)
my_data = np.random.randint(0,101,(4,3))
```

```python
my_data
```

```python
my_index = ['CA','NY','AZ','TX']
my_columns = ['Jan','Feb','Mar']
```

```python
df = pd.DataFrame(data=my_data)
df
```

```python
df = pd.DataFrame(data=my_data, index=my_index)
df
```

```python
df = pd.DataFrame(data=my_data, index=my_index, columns=my_columns)
df
```

```python
df.info()
```
